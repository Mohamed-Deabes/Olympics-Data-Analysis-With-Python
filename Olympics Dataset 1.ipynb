```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes = pd.read_csv('Downloads/athlete_events.csv')
regions = pd.read_csv('Downloads/noc_regions.csv')
athletes.head()
```

```python
regions.head()
```

```python
# join dataframes
athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns = {'region':'Region', 'notes':'Notes'},inplace = True);
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
# check null values
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# top countrties participiting
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending =False).head(10)
top_10_countries
```

```python
fig = plt.figure(figsize =(12,6))
# plt.xticks(rotation = 20)
plt.title('overall participation by country')
sns.barplot(x =top_10_countries.index, y=top_10_countries, palette = 'Set2');
```

```python
# number age of particpition
fig = plt.figure(figsize=(12,6))
plt.title('Age distribution of participition')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor = 'black');
```

```python
# winter olympics sports
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer olympics sports
Summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
Summer_sports
```

```python
# n of male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# pie plot for male and female
fig = plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels = gender_counts.index, autopct='%1.1f%%', startangle = 180, shadow = True);
```

```python
#total medals
medals_counts = athletes_df.Medal.value_counts()
medals_counts
```

```python
# female-participations
female_participants = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season =='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
women_olympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season =='Summer')]
```

```python
sns.set(style='darkgrid')
fig = plt.figure(figsize=(20,10))
sns.countplot(x = 'Year', data = women_olympics, palette ="Spectral")
plt.title('Women Participants')
```

```python
part = women_olympics.groupby('Year')['Sex'].value_counts()
fig = plt.figure(figsize=(20,10))
part.loc[: ,'F'].plot()
plt.title('plot of women athletes in olympics')
```

```python
totalgoldmedals =goldmedals.Region.value_counts().reset_index(name='Medal').head(6)
g =sns.catplot(x="index", y="Medal", data=totalgoldmedals, height =5, kind ="bar", palette = "rocket")
g.despine(left = True)
g.set_xlabel("top countries")
g.set_ylabel("number of medals")
plt.title('Most Countries With Gold Medal')
```
